### **bar plot**

```python
days_of_week = ['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
time_spent = [10, 15, 21, 13, 35, 80, 60]

fig , ax = plt.subplots(figsize=(10, 4))
ax.bar(days_of_week, time_spent)
ax.set_title("time spent on instagram")
ax.set_xlabel("days")
ax.set_ylabel("minutes")
```

**horizontal barplot**

```python
ax.clear()
ax.barh(days_of_week, time_spent)
fig
```

**sort per time spent**

```python
days_of_week = np.array(['Saturday', 'Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'])
time_spent = np.array([10, 15, 21, 13, 35, 80, 60])
```

```python
mask = time_spent.argsort()
days_of_week[mask]
```

```python
days_of_week = days_of_week[mask]
time_spent.sort()
```

**sorted barplot**

```python
ax.clear()
ax.bar(days_of_week, time_spent)
ax.set_title("time spent on instagram")
ax.set_xlabel("days")
ax.set_ylabel("minutes")
fig
```

## **seaborn**

### **regplot**

```python
reg_plot = sns.regplot(data=houses_df, x='area', y='price')
```

```python
dis_plot = sns.displot(houses_df, x='price', bins = 13, kde=True)
```

### **heatmap**

```python
# numeric columns
numeric_columns = houses_df.select_dtypes(include=[np.number]).columns
numeric_columns

# heatmap
sns.heatmap(houses_df[numeric_columns].corr(), annot=True, cmap='coolwarm')
```

```python
fig = px.scatter(houses_df, x='area', y='price', width=800, height=600, color='bedrooms', size='parking', title='House Price vs. Area') # color_continuous_scale=px.colors.sequential.Viridis
fig.update_layout(
    title='House Price vs. Area',
    xaxis_title='Area',
    yaxis_title='Price'
)

    
fig.show()
```
